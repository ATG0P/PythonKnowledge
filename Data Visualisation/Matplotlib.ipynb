```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

MATPLOTLIB BASIC

```python
x = np.linspace(0,5,11)#a 1D array whose elements are bwn 0 and 5(inclusive) of length 11
x
y=x**3
y
```

```python
plt.plot(x,y)
plt.title("Y vs X cubed Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

SUBPLOTS IN MATPLOTLIB

```python
plt.subplot(2,2,1)
plt.plot(x,x)
plt.subplot(2,2,2)
plt.plot(y,x)
plt.subplot(2,2,4)
plt.plot(x,y**(1/4))
plt.subplot(2,2,3)
plt.plot(x,y)
```

OBJECT ORIENTED WAY

```python
#This is used to access the canvas
fig = plt.figure()
fig
```

```python
fig.add_axes([0.1,0.1,0.8,0.8])
#(10% away from left side, 10% away from bottom, 80% width, 80% height)
fig
```

```python
fig1 = plt.figure()
a1= fig1.add_axes([0.1,0.1,0.8,0.8])
a1.plot(x,y)
a2=fig1.add_axes([0.1,0.2,0.6,0.6])
a2.plot(x,y)#they overlap
```

```python
fig2 = plt.figure(figsize=(7,9),dpi=150)#figsize is canvas size = 700 x 900 (By Default it is 640 x 800)
#more the dpi, clearer the pic(dpi=100 by default)
k=fig2.add_axes([0.1,0.1,0.8,.8])
k.plot(x,y)
```

TYPES OF PLOTS

```python
#scatter plots 
plt.scatter(x,y)
```

```python
#histogram (takes only one argument btw)
plt.hist(x)
```

```python
#boxplot
plt.boxplot(x)
```

SAVING PLOTS

```python
fig4=plt.figure()
ax = fig4.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("Y_vs_X_Cubed.png")
```

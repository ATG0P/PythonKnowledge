```python
def wicketexracter(w):
    r=''
    for i in range(0,len(w)-1):
        if w[i]=='-' and w[i+1]=='-':
            r=w[i-1]
        else:
            continue
    r=int(r)
    return r
```

```python
df['highest_wickets']=df['best_bowling_figure'].apply(wicketexracter)
df['highest_wickets']
```

```python
top_bowlers=df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head(10)
top_bowlers.plot(kind='bar')
```

```python
def figextracter(f):
    r = ' '
    for i in range(0,len(f)):
        if f[i]=='-' and f[i+1]=='-':
            for j in range(i+2,len(f)):
                r+=f[j]
    r=int(r)
    return r
```

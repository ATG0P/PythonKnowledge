```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df
```

```python
sns.countplot(x=df["sex"],hue=df['smoker'])
```

```python
import numpy as np
sns.barplot(x=df['sex'],y=df['total_bill'],estimator=np.sum)#by default is mean
#Also, you can use estimator = sum, but it is slower and less professional 
```

```python
sns.boxplot(x='tip',y='day',data=df)
```

```python
sns.violinplot(x='tip',y='day',data=df)
```

```python
sns.stripplot(x='day',y='total_bill',data=df)
```

```python
sns.swarmplot(x='day',y='total_bill',data=df)#the congested data get's a little distributed
```

```python
sns.violinplot(x='day',y='total_bill',data=df)
sns.swarmplot(x='day',y='total_bill',data=df)
```

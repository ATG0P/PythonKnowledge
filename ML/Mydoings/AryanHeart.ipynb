```python
df_cleaned.dtypes
```

```python
df_cleaned['Sex'].value_counts()
```

```python
df_cleaned['Sex'] = df_cleaned['Sex'].map({
    'M':1,
    'F':0
})
df_cleaned.head()
```

```python
df_cleaned.rename(columns={
    'Sex':'is_male'
},inplace=True)
df_cleaned.head()
```

```python
df_cleaned['ChestPainType'].value_counts()
```

```python
df_cleaned = pd.get_dummies(df_cleaned,columns=['ChestPainType'])
df_cleaned.head()
```

```python
df_cleaned = pd.get_dummies(df_cleaned,columns=['RestingECG','ST_Slope'])
df_cleaned.head()
```

```python
import numpy as np
import seaborn as sns
from    scipy.stats import chi2_contingency
import pandas as pd
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

<h1>LET'S FIND THE ASSOCIATION OF GENDERS WITH SURVIVAL

```python
contigencytable = pd.crosstab(df['sex'],df['survived'])
contigencytable
```

```python
chi2, p_value, dof, expected = chi2_contingency(contigencytable)
```

```python
chi2#chi squared valued
```

```python
p_value
```

```python
dof#(2-1)(2-1)=1 x 1 = 1
```

```python
expected
```

```python
alpha=0.05
```

```python
if p_value < alpha:
    print("we reject the null hypothesis and there is a significant relationship between gender and survivl")
else:
    print("we accept the null hypothesis there is no connenction")
```

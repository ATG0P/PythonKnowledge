```python
import numpy as np
from scipy import stats
```

```python
group_A = [85, 88, 90, 92, 87, 85, 89, 91, 86, 88] 
group_B = [82, 84, 80, 83, 81, 79, 78, 85, 84, 83]
```

```python
na = len(group_A)
nb = len(group_B)
```

```python
ma = np.mean(group_A)
mb = np.mean(group_B)
```

```python
sa = np.std(group_A)
sb = np.std(group_B)
```

```python
t = (ma-mb)/np.sqrt(((sa*sa)/na)+((sb*sb)/nb))
t
```

```python
p_value = 2*(1-stats.t.cdf(abs(t),df=na-1))
alpha = 0.05
```

```python
if p_value < alpha:
    print("reject the null hypothesis")
else:
    print("accept the null hypothesis")
```

```python
t_stats, p_value = stats.ttest_ind(group_A,group_B, equal_var=False)
```

```python
alpha = 0.05
```

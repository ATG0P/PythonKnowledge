```python
import numpy as np
from scipy import stats
```

```python
sample = [172, 174, 168, 169, 171, 173, 175, 170, 169, 172]
sample_mean = np.mean(sample)
sample_mean
```

```python
sample_std = np.std(sample,ddof=1) #here ddof = degree of freedom and it's 1 cuz it's a t test
sample_std
```

```python
n = len(sample)
population_mean = 170
t_value = (sample_mean-population_mean)/(sample_std/np.sqrt(n))
t_value
```

```python
p_value = 2*(1-stats.t.cdf(abs(t_value),df=n-1))
alpha = 0.05
```

```python
if p_value < alpha:
    print("reject the null hypothesis")
else:
    print("accept the null hypothesis")
```

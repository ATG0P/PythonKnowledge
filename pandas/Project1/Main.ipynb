```python
#task 2 - learning feature extracion
def episode_extracter(str):
    s=str.index('(')
    e=str.index(' eps)')
    num=""
    for i in range (s+1,e+1):
        num+=str[i]
    num=int(num)
    return num
```

```python
df["Episodes"]=df["Title"].apply(episode_extracter)
df.head()
```

```python
def time_stamp_extracter(txt):
    s=txt.find(")")
    e=s+19
    r=""
    for i in range(s+1,e+1):
        r+=txt[i]
    return r
```

```python
df["Time Stamp"]=df["Title"].apply(time_stamp_extracter)
df.head()
```

```python
def year_extracter(txt):
    s=txt.find(")")
    e=s+19
    y1=""
    y2=""
    for i in range(s+1,e+1):
        if (txt[i].isdigit() and i<=s+8):
            y1+=txt[i]
        elif(txt[i].isdigit() and i>s+8):
            y2+=txt[i]
        else:
            continue
    y1=int(y1)
    y2=int(y2)
    return y2-y1
```

```python
df["Year Gap"]=df['Title'].apply(year_extracter)
df[df["Year Gap"]==df["Year Gap"].max()]
```

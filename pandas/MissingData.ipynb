```python
import numpy as np 
import pandas as pd
```

```python
#nan=not a number(i.e; null value)
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
```

```python
df
```

Finding missing Data

```python
#to find if a column has a nan value
df.isna()
```

```python
#to find the number of nan values of rows:
df.isna().sum()
```

```python
#to check if every column contains nan value or not
df.isna().any()
```

Removing any NaN data row

```python
df.dropna()
```

```python
#to print rows which has n number of NON-NULL values
df.dropna(thresh=3)
```

```python
df.dropna(thresh=4)
```

```python
df.dropna(thresh=5)
```

Filling the missing data

```python
#to fill all the nan values with same number:-
df.fillna(25)
```

But what if I want to fill a specific number for nan value for a specific column?

```python
fillnavalues = {'A':0,'B':25,'C':100,'D':1000}
df.fillna(value=fillnavalues)
```

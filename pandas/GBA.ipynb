GBA = GROUP BY AGGREGATION

```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}
df = pd.DataFrame(data)
df
```

```python
cat=df.groupby("Category")#here cat became an object
for a,b in cat:
    print(a)
    print(b)    
```

```python
#QUESTION: Group the category data and find the sum of the sales
sumaddsales = df.groupby("Category")["Sales"].sum()
sumaddsales
```

```python
#QUESTION: Group by store and calculate the sum of sales
sumaddstores=df.groupby("Store")["Sales"].sum()
sumaddstores
```

```python
cat=df.groupby(["Category","Store"])["Sales"].sum()
cat
```

Aggregation means sum,mean,min,max,count std operation

```python
df["Sales"].agg(["mean","median","sum","min","max","std","count","std"])
```

```python
import numpy as np
import pandas as pd
```

```python
#Dataframes are multi dimension series
```

```python
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}
```

```python
df = pd.DataFrame(data)#or just type pd.DataFrame(data) and you'll still get the output on the notebook  
print(df)
```

```python
data = [
    ["Aryan",18],
    ["Anjali",14],
    ["Angad",15]
]
pd.DataFrame(data)
```

```python
column=["Name","age"]
pd.DataFrame(data,columns=column)
```

```python
df2 = pd.DataFrame(data,columns=column)
df2["Name"] #will only select Name column
```

```python
df2[["Name","age"]]#to show multiple columns at once, you'll have to put in a list of columns
```

```python
df2["Dream"]=["AIML Engineer", "Stock Broker", "Boxer"]
df2["Gender"]=["M","F","M"]
df2
```

Now to remove a column, Let's Say: the 'age' column:-

```python
df2.drop("age",axis=1)
```

```python
df2
```

Notice how it doesn't permanently deletes the "age" column? to do that...

```python
df2.drop("age",axis=1,inplace=True)
df2
```

To delete multiple columns...

```python
df2.drop(["Gender","Dream"],axis=1)#not using implace = True because I didn't wanted to mess up the table
```

```python
df2
```

To delete a row..

```python
df2.drop(0,axis=0)#not required to type axis=0 as it is by default(note: it's not permanently deleted yet unless you type: inplace=True)
```

Now to access a row....

```python
df2.loc[0]#first element say
```

To access multiple row...

```python
df2.loc[[0,1]]
```

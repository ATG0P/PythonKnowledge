MJC = MERGING JOINING CONCANTINATION

```python
import numpy as np
import pandas as pd
```

```python
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

# DataFrame 2: Salary information
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees,salaries)
```

```python
pd.merge(employees,salaries,on="employee_id",how="inner")#same as above(basically default parameter of above cell)
```

```python
m=pd.merge(employees,salaries,on="employee_id",how="outer")#merges the table on the basis of emloyee id, and taking outer rows though they are not common employee ids
m
```

```python
pd.merge(employees,salaries,on="employee_id",how="left")#merge on the basis of employee_id wrt left parameter (here employees)
```

```python
pd.merge(employees,salaries,on="employee_id",how="right")#merge on the basis of employee_id wrt right parameter (here salaries)
```

CONCATINATION OF 2 DATAFRAMES

```python
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
    'C': ['C0', 'C1', 'C2']
})

df2 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5'],
    'C': ['C3', 'C4', 'C5']
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])
```

```python
#what if I want to concatinate on the basis of rows?
pd.concat([df1,df2],axis=1)
```

**JOINING 2 DATAFRAMES**


```python
df1 = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie']
}, index=[1, 2, 3])

# Second DataFrame
df2 = pd.DataFrame({
    'score': [85, 90, 75]
}, index=[2, 3, 4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2)
```

```python
df1.join(df2,how="outer")
```

```python
pd.concat([df1,df2])
```

```python
import numpy as np
import pandas as pd
```

```python
#series = 1D list which can store multiple data type
#labels = indexes
labels=['a','b','c']
my_list=[10,20,30]
arr=np.array([10,20,30])
d={"one":10,"two":20,"three":30}
```

```python
pd.Series(my_list)#will print all the values of list 'my_list' alongside indexes
```

```python
pd.Series(my_list,index=labels)
#will print all the values of list 'my_list' with indexes being a,b and c
```

```python
pd.Series(arr)
```

```python
pd.Series(d)
```
